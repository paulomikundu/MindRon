```python
from sentence_transformers import SentenceTransformer

# Define the paragraph
paragraph = """
Aarav stood at the edge of the old bookstore, his fingers brushing against the weathered wooden door. The sign above, once golden and bright, now hung crookedly, its letters faded with time. "The Last Page"—a name that once held magic, a place that once felt like home.

He hesitated. It had been over a decade since he last set foot in this shop, yet nothing else in the city had ever held the same sense of warmth, of quiet wonder. The world outside had changed—glass towers rose where parks once stood, neon signs blinked in a rhythm that never stopped, and life moved at a relentless pace. But inside this shop, time had always seemed to stand still.

With a deep breath, he pushed the door open. A bell above jingled, its chime soft yet persistent, as if welcoming him back after all these years. The scent of old books and parchment filled his lungs, mingling with the faint aroma of ink and dust. Sunlight filtered through a cracked window, illuminating floating dust particles like tiny golden stars.

His heart pounded.

This wasn't just any bookstore for him—it was where he had spent countless hours as a child, curled up in the corner while his mother read aloud to him, weaving stories of lost kingdoms, forgotten heroes, and whispers of ancient magic.

And now, he was back—not for nostalgia, not for the books—but for something much deeper.

Behind the wooden counter, an old man sat hunched over a thick ledger, spectacles perched on the bridge of his nose. He hadn't changed much—his hair was grayer, perhaps, his shoulders slightly more stooped—but his presence still carried the same quiet authority.

The old man glanced up, his sharp eyes softening the moment they landed on Aarav.

"Back again, are you?" he said, voice rough yet familiar.

Aarav swallowed. "I need to find a book."

Mr. Das chuckled, closing his ledger with a practiced motion. "That’s what they all say."

"No, I mean… a specific book." Aarav hesitated, his voice faltering. "My mother’s book."

For a moment, silence hung between them. Then Mr. Das's expression shifted from amusement to something deeper—understanding.

"I remember your mother well," he said quietly. "She used to sit by that window, reading for hours. A brilliant woman."

Aarav nodded. "She used to read me a story. About a lost kingdom… and a boy who could speak to shadows."

Mr. Das exhaled, his gaze distant as if he, too, was searching the past.

"What was the book’s name?" he asked.

Aarav’s throat tightened. That was the problem. He didn't know. He only remembered the words, the way his mother’s voice carried them, the rhythm of her storytelling like a lullaby. It was a part of him, a memory so deeply embedded that he had never thought to question the book’s existence—until now.

"I don’t know," he admitted. "I just know it was hers."

Mr. Das studied him for a long moment. Then, without another word, he stood, his footsteps soft as he moved toward the back of the shop. Aarav watched as he disappeared behind the towering shelves, his fingers grazing book spines with practiced ease.

Minutes passed, the only sounds the ticking of a clock and the occasional rustle of pages.

Then, Mr. Das returned.

In his hands, he held a leather-bound book, its spine cracked with age, its edges worn with time.

"This was hers," he said simply, offering it to Aarav.

Aarav took it with trembling fingers. The cover was smooth yet fragile, as though it had been touched a thousand times before. He flipped it open, and there, nestled between the pages, was something unexpected—a single, folded piece of paper.

He picked it up carefully, unfolding it with reverence. His mother’s handwriting—elegant and flowing—spilled across the parchment.

"When you are lost, remember—the story never ends. It only waits for you to turn the page."

A lump formed in Aarav’s throat.

Mr. Das spoke softly. "She left it here for you. She told me one day you'd come looking for it."

Aarav clenched his jaw, emotions swelling inside him like a tide. He had spent years running—running from the memories, from the grief, from the ache of loss that never truly faded. He had buried himself in work, in responsibilities, in distractions. But here, in the quiet sanctuary of "The Last Page," the past had finally caught up with him.

He traced his mother’s handwriting with his thumb, the ink slightly smudged, as if touched by time itself. He remembered her voice, her laughter, the way she always smelled like jasmine and old books.

He had thought he was searching for a book.

But in truth, he had been searching for her.

Aarav looked up, meeting Mr. Das’s gaze. "Thank you," he whispered.

The old man simply nodded. "Take your time."

Aarav exhaled slowly and moved toward the window where his mother once sat. The world outside continued rushing forward—people hurrying, cars honking, the city’s pulse never ceasing.

But inside this little bookstore, time had stopped.

And for the first time in years, Aarav allowed himself to sit, to breathe, to remember.

He opened the book, and as he turned the first page, he smiled.

The story was waiting.

And so was he."""

# Load pre-trained model
model = SentenceTransformer('sentence-transformers/paraphrase-MiniLM-L6-v2')

# Split paragraph into sentences
sentences = paragraph.split(". ")

# Encode the sentences
embeddings = model.encode(sentences)

# Print the embeddings
for i, sentence in enumerate(sentences):
    print(f"Sentence {i+1}: {sentence}")
    print(f"Embedding: {embeddings[i]}\n")
```

```python
from transformers import pipeline

# Load the question-answering model
qa_pipeline = pipeline("question-answering", model="ixa-ehu/SciBERT-SQuAD-QuAC")

# Take user input for the question
question = input("Enter your question: ")

# Get the answer from the model
answer = qa_pipeline(question=question, context=paragraph)

# Print the result
print(f"\nQuestion: {question}")
print(f"Answer: {answer['answer']}")
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the tokenizer and model
tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-large")
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-large")

# Take user input for context and question
context = input("Enter the context: ")
question = input("Enter your question: ")

# Format input as "question: <question> context: <context>"
input_text = f"question: {question} context: {context}"

# Tokenize the input
input_ids = tokenizer(input_text, return_tensors="pt").input_ids

# Generate the answer
outputs = model.generate(input_ids)

# Decode and print the answer
answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(f"\nContext: {context}")
print(f"Question: {question}")
print(f"Answer: {answer}")
```

```python
from transformers import pipeline

# Load the zero-shot classifier
classifier = pipeline("zero-shot-classification", model='cross-encoder/nli-deberta-v3-base')

# Take story input
print("Enter your short story (Press Enter twice when done):")
story_lines = []
while True:
    line = input()
    if line == "":
        break
    story_lines.append(line)

story = "\n".join(story_lines)  # Combine lines into a full story

# Define detailed genre labels for better classification
candidate_labels = [
    "high fantasy", "urban fantasy", "science fiction", "detective mystery",
    "psychological thriller", "romance drama", "paranormal horror",
    "historical fiction", "adventure", "mythology", "dystopian fiction"
]

# Prepend a structured prompt to improve classification accuracy
prompt_story = f"The following text is a story. Identify its most appropriate genre:\n\n{story}"

# Perform classification
result = classifier(prompt_story, candidate_labels)

# Get the most probable category and confidence score
predicted_category = result['labels'][0]
confidence_score = result['scores'][0]

# If confidence is below 90%, refine labels and retry
if confidence_score < 0.90:
    print("\nLow confidence detected. Refining genre classification...")

    # Select the top 3 highest-scoring categories
    refined_labels = result['labels'][:3]

    # Retry classification with refined labels
    refined_result = classifier(prompt_story, refined_labels)
    predicted_category = refined_result['labels'][0]
    confidence_score = refined_result['scores'][0]

# Print the results
print(f"\nPredicted Story Type: {predicted_category} (Confidence: {confidence_score:.2f})")
print("\nFull classification results:")
for label, score in zip(result['labels'], result['scores']):
    print(f"{label}: {score:.2f}")
```

```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer, pipeline

# Load the summarization model
model_name = "google/pegasus-xsum"
tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name)

# Load the emotion classification pipeline
emotion_classifier = pipeline("text-classification", model="bhadresh-savani/distilbert-base-uncased-emotion")

# Take story input
print("Enter your story (Press Enter twice when done):")
story_lines = []
while True:
    line = input()
    if line == "":
        break
    story_lines.append(line)

story = "\n".join(story_lines)  # Combine lines into a full story

# Take summary length inputs
min_length = int(input("Enter minimum summary length: "))
max_length = int(input("Enter maximum summary length: "))

# Generate summary
inputs = tokenizer(story, return_tensors="pt", max_length=1024, truncation=True)
summary_ids = model.generate(inputs["input_ids"], max_length=max_length, min_length=min_length, length_penalty=2.0)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Detect emotion from summary
emotion_result = emotion_classifier(summary)[0]
emotion_label = emotion_result["label"]
emotion_score = emotion_result["score"]

# Print results
print("\nGenerated Summary:\n", summary)
print(f"\nDetected Emotion: {emotion_label} (Confidence: {emotion_score:.2f})")
```

```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer, pipeline

# Load Summarization Model
summarizer_model = "google/pegasus-xsum"
summarizer_tokenizer = PegasusTokenizer.from_pretrained(summarizer_model)
summarizer = PegasusForConditionalGeneration.from_pretrained(summarizer_model)

# Load Classifiers
genre_classifier = pipeline("zero-shot-classification", model='cross-encoder/nli-deberta-v3-base')
emotion_classifier = pipeline("text-classification", model="bhadresh-savani/distilbert-base-uncased-emotion")

# Take Story Input
print("Enter your story (Press Enter twice when done):")
story_lines = []
while True:
    line = input()
    if line == "":
        break
    story_lines.append(line)

story = "\n".join(story_lines)  # Combine lines into a full story

# Validate Input
if len(story.strip()) == 0:
    raise ValueError("Error: No story input detected. Please enter a valid story.")

# Define possible story genres
candidate_labels = ["fantasy", "science fiction", "mystery", "romance", "horror", "adventure", "historical fiction"]

# Classify Story Type
classification_result = genre_classifier(story, candidate_labels)
predicted_genre = classification_result['labels'][0]
confidence_genre = classification_result['scores'][0]

# Take Summary Length Inputs
min_length = int(input("Enter minimum summary length: "))
max_length = int(input("Enter maximum summary length: "))

# Tokenize Story for Summarization
inputs = summarizer_tokenizer(story, return_tensors="pt", max_length=1024, truncation=True)
token_count = inputs["input_ids"].shape[1]

# Adjust min/max length if they exceed available tokens
if min_length >= token_count:
    min_length = max(10, token_count // 2)  # Set reasonable minimum
if max_length >= token_count:
    max_length = max(min_length + 10, token_count - 1)

# Generate Summary
summary_ids = summarizer.generate(inputs["input_ids"], max_length=max_length, min_length=min_length, length_penalty=2.0)
summary = summarizer_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Detect Emotion from Summary
emotion_result = emotion_classifier(summary)[0]
emotion_label = emotion_result["label"]
emotion_score = emotion_result["score"]

# Print Results
print("\n" + "="*30 + " Story Analysis " + "="*30)
print(f"📖 Predicted Story Genre: {predicted_genre} (Confidence: {confidence_genre:.2f})")

print("\n" + "="*30 + " Generated Summary " + "="*30)
print(summary)

print("\n" + "="*30 + " Detected Emotion " + "="*30)
print(f"😊 Emotion: {emotion_label} (Confidence: {emotion_score:.2f})")
```
